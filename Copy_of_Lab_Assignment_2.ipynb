```python
# M.Tech CSE Weekend Program - Assignment 2
# Understanding Python Data Structures Through Real-World Applications with Visualization

# ---
## **ðŸ“Œ Case Study 1: Analyzing Student Performance**

# Step 1: Without Using Lists
# Storing marks of 10 students and finding the average
marks1 = 75
marks2 = 80
marks3 = 90
marks4 = 85
marks5 = 95
marks6 = 70
marks7 = 60
marks8 = 55
marks9 = 65
marks10 = 88

average = (marks1 + marks2 + marks3 + marks4 + marks5 + marks6 + marks7 + marks8 + marks9 + marks10) / 10
print("Average Marks:", average)
```

```python
# Step 2: Using a List
marks = [75, 80, 90, 85, 95, 70, 60, 55, 65, 88]
average = sum(marks) / len(marks)
print("Average Marks:", average)
```

```python
# Visualization: Histogram
import matplotlib.pyplot as plt
plt.hist(marks, bins=5, color='blue', edgecolor='black')
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.title("Student Marks Distribution")
plt.show()
```

```python
## **ðŸ“Œ Case Study 2: Finding Common Students in Two Courses**

math_students = {"Alice", "Bob", "Charlie", "David", "Eve"}
science_students = {"Charlie", "David", "Eve", "Frank", "Grace"}

common_students = math_students.intersection(science_students)
print("Common Students:", common_students)
```

```python
## **ðŸ“Œ Case Study: Analyzing Sales Data of an Electronics Store**

# **Problem Context:**
# A retail store sells different electronic products like laptops, phones, tablets, and accessories.
# The store manager wants to analyze the sales data of the last 30 days to:
# - Identify the best-selling products
# - Find the total revenue generated
# - Analyze the sales trend
# - Identify which product category needs more promotions

# **Given Sales Data (Dictionary format)**
sales_data = {
    "Laptop": [12, 15, 20, 18, 25, 22, 19, 16, 18, 21, 24, 27, 30, 33, 25, 28, 26, 31, 35, 29, 27, 22, 24, 20, 18, 16, 15, 12, 10, 8],
    "Phone": [50, 48, 60, 55, 58, 62, 65, 67, 70, 75, 80, 85, 90, 85, 78, 73, 70, 68, 65, 60, 58, 55, 53, 50, 48, 45, 42, 40, 38, 35],
    "Tablet": [30, 28, 35, 33, 32, 30, 28, 25, 22, 20, 18, 15, 14, 12, 10, 9, 8, 7, 6, 5, 6, 8, 10, 12, 14, 15, 17, 18, 19, 20],
    "Accessories": [80, 85, 88, 90, 92, 95, 98, 100, 105, 110, 115, 120, 125, 128, 130, 135, 140, 145, 150, 155, 160, 162, 165, 170, 175, 180, 185, 190, 195, 200]
}

# **Step 1: Identify the Best-Selling Product**
total_sales = {product: sum(sales) for product, sales in sales_data.items()}
best_selling = max(total_sales, key=total_sales.get)
print("Best-Selling Product:", best_selling)

# **Step 2: Calculate Total Revenue**
unit_prices = {"Laptop": 75000, "Phone": 45000, "Tablet": 30000, "Accessories": 5000}
total_revenue = sum(total_sales[product] * unit_prices[product] for product in sales_data)
print("Total Revenue Generated:", total_revenue)

# **Step 3: Visualizing Sales Trends**
import matplotlib.pyplot as plt
for product, sales in sales_data.items():
    plt.plot(range(1, 31), sales, marker='o', linestyle='-', label=product)
plt.xlabel("Day")
plt.ylabel("Units Sold")
plt.title("Daily Sales Trend of Products")
plt.legend()
plt.show()

# **Step 4: Bar Chart of Total Sales**
plt.bar(total_sales.keys(), total_sales.values(), color=['blue', 'green', 'red', 'purple'])
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.title("Total Sales of Each Product")
plt.show()

# ---
## **ðŸ”¹ Practice Questions**
# 1ï¸âƒ£ Based on the given sales data, which product had the lowest total sales? Modify the code to find this.
# 2ï¸âƒ£ Compute and visualize the percentage contribution of each product to the total revenue using a pie chart.
# 3ï¸âƒ£ Find the day on which the highest number of any product was sold. Identify the product and the day.
# 4ï¸âƒ£ Suggest promotional strategies based on the trends observed in the sales data.
```
